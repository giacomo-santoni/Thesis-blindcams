```python
file_drdf_list = ["/Users/giacomosantoni/Desktop/TESI/Progetto_ML/blindcams/data/good_data_last4/response.drdf"]
file_root_list = ["/Users/giacomosantoni/Desktop/TESI/Progetto_ML/blindcams/data/good_data_last4/sensors.root"]
```

```python
images = prep.PreprocessNotScaled(file_drdf_list[0])
root_images = rooprep.RootPreprocessing(file_root_list[0], file_drdf_list[0])
```

```python
_, blind_ev = prep.FindBlindEvents(root_images,images)
_, not_blind_ev = prep.FindNotBlindEvents(root_images,images)
```

```python
all_means = []
all_max = []
for i in images:
    mean = np.mean(i)
    max = np.max(i)
    all_means.append(mean)
    all_max.append(max)
```

```python
ev_list = np.arange(0,len(images))
```

```python
plt.scatter(ev_list,all_means)
plt.show()
```

```python
#MEAN
blind_mean = []
for i in blind_ev:
    mean = np.mean(i)
    blind_mean.append(mean)

not_blind_mean = []
for i in not_blind_ev:
    mean = np.mean(i)
    not_blind_mean.append(mean)
```

```python
#MAX
blind_max = []
for i in blind_ev:
    max = np.max(i)
    blind_max.append(max)

not_blind_max = []
for i in not_blind_ev:
    max = np.max(i)
    not_blind_max.append(max)
```

```python
#MEDIAN
blind_median = []
for i in blind_ev:
    median = np.median(i)
    blind_median.append(median)

not_blind_median = []
for i in not_blind_ev:
    median = np.median(i)
    not_blind_median.append(median)
    
```

```python
ev_blind_list = np.arange(0,len(blind_ev))
ev_not_blind_list = np.arange(0,len(not_blind_ev))
```

```python
plt.scatter(blind_mean, ev_blind_list, label='blind')
plt.scatter(not_blind_mean, ev_not_blind_list, label='not blinded')
plt.legend(loc='upper right') 
#plt.xlim(0,200)
plt.show()

plt.scatter(blind_max, ev_blind_list, label='blind')
plt.scatter(not_blind_max, ev_not_blind_list, label='not blinded')
plt.legend(loc='upper right') 
#plt.xlim(0,200)
plt.show()

plt.scatter(blind_median, ev_blind_list, label='blind')
plt.scatter(not_blind_median, ev_not_blind_list, label='not blinded')
plt.legend(loc='upper right') 
#plt.xlim(0,200)
plt.show()
```

```python
plt.scatter(blind_median, blind_mean, label='blind')
plt.scatter(not_blind_median, not_blind_mean, label='not blinded')
plt.legend(loc='upper right') 
#plt.xlim(0,200)
# plt.xscale('log')
# plt.yscale('log')
plt.show()

plt.scatter(blind_median, blind_max, label='blind')
plt.scatter(not_blind_median, not_blind_max, label='not blinded')
plt.legend(loc='upper right') 
#plt.xlim(0,200)
# plt.xscale('log')
# plt.yscale('log')
plt.show()

plt.scatter(not_blind_max, not_blind_mean, label='not blinded')
plt.scatter(blind_max, blind_mean, label='blind')
plt.legend(loc='upper right') 
plt.ylim(0,250)
plt.xlim(0,1000)
# plt.xscale('log')
# plt.yscale('log')
plt.show()
```
